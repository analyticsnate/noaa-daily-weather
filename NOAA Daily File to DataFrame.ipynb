```python
import pandas as pd
```

```python
# set initial column values
colsList = ['ID','YEAR','MONTH','ELEMENT']

# set initial column widths
widthsList = [11,4,2,4]
```

```python
# add VALUE(n), MFLAG(n), QFLAG(n), and SFLAG(n) for each day of the month
for n in range(1,32):
    
    # append column names to colsList
    colsList.append('VALUE{}'.format(n))
    colsList.append('MFLAG{}'.format(n))
    colsList.append('QFLAG{}'.format(n))
    colsList.append('SFLAG{}'.format(n))
    
    # append column widths to widthsList
    widthsList.append(5)
    widthsList.append(1)
    widthsList.append(1)
    widthsList.append(1)
    
```

```python
df = pd.read_fwf('USW00014942.dly',header=None,#delimiter=' ',
                widths=widthsList,names=colsList,index_col=False)
#df = df.reset_index()
```

```python
# YEAR = 2018, ELEMENT = TMAX (max temp), 
# showing columns MONTH, 
# VALUE1 (value on 1st day of month)
# VALUE2 (value on 2nd day of month)
# VALUE 29 (value on 29th day of month)
# values are in C * 10, so -177 is -17.7 C, which is 0.14 F
df[(df['YEAR']==2018)&(df['ELEMENT']=='TMAX')].filter(['MONTH','VALUE1','VALUE2','VALUE29'])
```

```python
# build values list
valuesList = []
for i in range(1,32):
    valuesList.append('VALUE{}'.format(i))
```

```python
df_Values = df.filter(valuesList)
df_Values.head()
```

```python
df_Values_T = df_Values.transpose()
```

```python
df_Values_T.head()
```

```python
df_Values = pd.DataFrame(columns=['ID','YEAR','MONTH','ELEMENT','VALUE','DAY'])

for i in range(1,32):
    df_Append = df.filter(['ID','YEAR','MONTH','ELEMENT','VALUE{}'.format(i)])
    df_Append = df_Append.rename(columns={'VALUE{}'.format(i):'VALUE'})
    df_Append['DAY'] = i
    df_Values = df_Values.append(df_Append)
    
df_Values.head()
```

```python
df_Values.to_excel('Omaha_Tmax.xlsx') #[(df_Values['YEAR']==1948)&(df_Values['ELEMENT']=='TMAX')]
```
